```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**Import dataset**

```python
dataset = pd.read_csv("Big Basket Cart Prediction.csv", header= None)
transactions = []
for i in range(0, 7219):
  transactions.append([str(dataset.values[i,j]) for j in range(0,20)])


```

**Apriori training on dataset**

```python
from apyori import apriori
basket_intelligence = apriori( transactions= transactions, min_support = 0.002, min_confidence = 0.2, min_lift= 3, min_length= 2)
```

**Visualizing**

**Raw results**

```python
results= list(basket_intelligence)
```

```python
print(results)
```

**Display**

```python
def inspect(results):
  product1 =    [tuple(result[2][0][0])[0] for result in results]
  product2 =    [tuple(result[2][0][1])[0] for result in results]
  supports =    [result[1] for result in results]
  confidence =  [result[2][0][2] for result in results]
  lift =       [result[2][0][3] for result in results]
  return list(zip( product1, product2, supports, confidence, lift))
DataFrame_intelligence = pd.DataFrame(inspect(results), columns= ['Product1', 'Product2', 'supports', 'Confidence', 'Lift'])
```
